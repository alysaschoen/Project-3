```python
import pandas as pd
```

```python
#Import CSV and convert to Pandas DataFrame

airbnb_df = pd.read_csv('Airbnb_Open_Data.csv')
airbnb_df.head()
```

```python
#Clear out unnecessary coulmns

airbnb_df = airbnb_df.drop(['host id', 'host_identity_verified', 'host name', 'country', 'country code', 'availability 365',
                           'minimum nights', 'last review'], axis=1)
airbnb_df.head()
```

```python
# Check neighbourhood group' for any errors

hood_values = airbnb_df['neighbourhood group'].unique()
value_counts = airbnb_df['neighbourhood group'].value_counts()
print(hood_values)
print('-----------------')
print(value_counts)
```

```python
#Change 'brookln' to Brooklyn

brookln = airbnb_df.loc[airbnb_df['neighbourhood group']=='brookln']
airnbn_df = airbnb_df.loc[brookln.index, 'neighbourhood group'] = 'Brooklyn' 

#Change 'manhatan' to Manhattan

manhatan = airbnb_df.loc[airbnb_df['neighbourhood group']=='manhatan']
airnbn_df = airbnb_df.loc[manhatan.index, 'neighbourhood group'] = 'Manhattan' 
```

```python
#Make sure the values have been changed

new_hood_values = airbnb_df['neighbourhood group'].unique()
new_value_counts = airbnb_df['neighbourhood group'].value_counts()
print(new_hood_values)
print('-----------------------------------------------------------------------')
print(new_value_counts)
```

```python
#Change column names

airbnb_df = airbnb_df.rename(columns={
    'NAME': 'Name',
    'neighbourhood group': 'Borough',
    'neighbourhood': 'Neighbourhood',
    'lat': 'Lat',
    'long': 'Lon',
    'instant_bookable': 'Instant Booking',
    'cancellation_policy': 'Cancellation Policy',
    'room type': 'Room Type',
    'Construction year': 'Construction Year',
    'price': 'Price Per Night',
    'service fee': 'Service Fee',
    'number of reviews': 'Total Reviews',
    'reviews per month': 'Reviews Per Month',
    'review rate number': 'Review Rate Number',
    'calculated host listings count': 'Host Listings Count',
    'house_rules': 'House Rules',
    'license': 'License'
})

airbnb_df.head()
```
